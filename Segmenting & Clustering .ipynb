Scraping Wikipedia Page with BeautifulSoup to Get Table

```python
#Import necessary libraries and Wikipedia Page
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests

website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(website_url,'lxml')
```

```python
#Scrape table from Wiki page
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
#Render scraped table as Pandas Dataframe df
df = pd.read_html(str(My_table))
df = df[0]
df
```

```python
# Drop rows with boroughs not assigned
df=df[df.Borough != 'Not assigned'].reset_index()
df
```

```python
#Combining multiple postal codes into one row
df=df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df
```
